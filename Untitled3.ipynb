```python
import numpy as np
import cv2
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

def load_and_preprocess_image_cv(image_path):
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError("Unable to load image. Please check the file path.")
    
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # Define color ranges for common flower colors
    color_ranges = [
        ([0, 100, 100], [10, 255, 255]),  # Red
        ([130, 100, 100], [170, 255, 255]),  # Purple
        ([20, 100, 100], [30, 255, 255]),  # Yellow
        ([90, 100, 100], [130, 255, 255]),  # Blue
        ([25, 40, 40], [85, 255, 255])  # Green
    ]
    
    mask = np.zeros(hsv.shape[:2], dtype=np.uint8)
    for lower, upper in color_ranges:
        lower = np.array(lower, dtype=np.uint8)
        upper = np.array(upper, dtype=np.uint8)
        mask |= cv2.inRange(hsv, lower, upper)
    
    result = cv2.bitwise_and(img, img, mask=mask)
    return img, result, mask

def extract_features_cv(mask):
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return None
    largest_contour = max(contours, key=cv2.contourArea)
    
    # Get rotated rectangle for better measurements
    rect = cv2.minAreaRect(largest_contour)
    box = cv2.boxPoints(rect)
    box = np.int0(box)
    
    # Calculate width and height
    width = rect[1][0]
    height = rect[1][1]
    
    # Ensure width is always the larger dimension
    if width < height:
        width, height = height, width
    
    # Calculate area and perimeter
    area = cv2.contourArea(largest_contour)
    perimeter = cv2.arcLength(largest_contour, True)
    
    # Extract features
    sepal_length = width
    sepal_width = height
    petal_length = np.sqrt(area)
    petal_width = area / perimeter
    
    return np.array([sepal_length, sepal_width, petal_length, petal_width])

def create_cnn_model():
    base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    predictions = Dense(3, activation='softmax')(x)  # 3 classes for iris species
    model = Model(inputs=base_model.input, outputs=predictions)
    
    for layer in base_model.layers:
        layer.trainable = False
    
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

def load_and_preprocess_image_cnn(image_path):
    img = load_img(image_path, target_size=(224, 224))
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    return preprocess_input(img_array)

def classify_iris_hybrid(image_path, svm_model, cnn_model, scaler, iris):
    # Traditional CV approach
    original, processed, mask = load_and_preprocess_image_cv(image_path)
    cv_features = extract_features_cv(mask)
    
    # CNN approach
    cnn_input = load_and_preprocess_image_cnn(image_path)
    
    # SVM classification
    if cv_features is not None:
        cv_features = (cv_features - np.mean(cv_features)) / np.std(cv_features)
        cv_features_scaled = scaler.transform(cv_features.reshape(1, -1))
        svm_prediction = svm_model.predict(cv_features_scaled)[0]
        svm_probabilities = svm_model.predict_proba(cv_features_scaled)[0]
        svm_confidence = np.max(svm_probabilities)
    else:
        svm_confidence = 0
        svm_prediction = -1  # Invalid prediction

    # CNN classification
    cnn_predictions = cnn_model.predict(cnn_input)
    cnn_class = np.argmax(cnn_predictions[0])
    cnn_confidence = np.max(cnn_predictions[0])

    # Adjust confidence thresholds
    svm_threshold = 0.4  # Lowered from 0.7
    cnn_threshold = 0.3  # Lowered from 0.7

    if cnn_confidence > cnn_threshold:
        final_class = iris.target_names[cnn_class]
        final_confidence = cnn_confidence
    elif cv_features is not None and svm_confidence > svm_threshold:
        final_class = iris.target_names[svm_prediction]
        final_confidence = svm_confidence
    else:
        final_class = "Unknown (Not confident enough to classify)"
        final_confidence = max(svm_confidence, cnn_confidence)

    # Visualization
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle(f"Hybrid Iris Classification: {final_class}", fontsize=16)
    
    axs[0, 0].imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    axs[0, 0].set_title("Original Image")
    axs[0, 0].axis('off')
    
    axs[0, 1].imshow(cv2.cvtColor(processed, cv2.COLOR_BGR2RGB))
    axs[0, 1].set_title("Processed Image (CV)")
    axs[0, 1].axis('off')
    
    axs[1, 0].imshow(mask, cmap='gray')
    axs[1, 0].set_title("Mask (CV)")
    axs[1, 0].axis('off')
    
    if cv_features is not None:
        axs[1, 1].bar(range(4), cv_features, tick_label=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
        axs[1, 1].set_title("Extracted Features (CV)")
        axs[1, 1].set_ylabel("Value")
    else:
        axs[1, 1].text(0.5, 0.5, "No CV features extracted", ha='center', va='center')
        axs[1, 1].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    print(f"Final Classification: {final_class}")
    print(f"Confidence: {final_confidence:.2f}")
    print(f"SVM Confidence: {svm_confidence:.2f}")
    print(f"CNN Confidence: {cnn_confidence:.2f}")
    
    return final_class, final_confidence

# Load Iris dataset and prepare SVM model
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
svm_model = SVC(probability=True, random_state=42)
svm_model.fit(X_train_scaled, y_train)

# Create and "train" CNN model (in reality, you'd need to train this on a large dataset of flower images)
cnn_model = create_cnn_model()

# Example usage
image_path = input("Enter the path to the flower image: ")
predicted_class, confidence = classify_iris_hybrid(image_path, svm_model, cnn_model, scaler, iris)
```
